```python
from datetime import datetime, timedelta
from datafuzz.generators import DatasetGenerator

generator = DatasetGenerator({
    'output': 'pandas',
    'schema': {
        'air_quality_index': range(0, 400, 20),
        'user_id': 'faker.user_name',
        'location': 'faker.location_on_land',
        'top_pollutant': ['PM2.5', 'PM10', 'O3', 'NO2', 'SO2', 'CO'],
        'top_pollutant_concentration': range(1, 100)
    },
    'num_rows': 100000,
    'start_time': datetime.now() - timedelta(days=20),
    'output': 'dataset',
    'increments': 'seconds',
})

generator.generate()

order_dataset = generator.to_output()
order_dataset.records.to_csv('data/air_quality.csv', index=False)
```
